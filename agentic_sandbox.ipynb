````python
import re
````

````python
def extract_and_run_code(ai_text):
  found_text = re.search(r'```python\n(.*?)\n```', ai_text, re.DOTALL)
  if found_text == None:
    return None, "Error: I couldn't find the ```python tags in your answer."
  pure_code = found_text.group(1)
  safe_memory = {}
  try:
    exec(pure_code, {}, safe_memory)
    if 'result' in safe_memory:
      final_answer = safe_memory['result']
      return final_answer, None
    else:
      return None, "Error: You forgot to save your answer in a variable named 'result'."
  except Exception as crashed_err:
    err_name = type(crashed_err).__name__
    err_reason = str(crashed_err)
    return None, f"Your code crashed! {err_name}: {err_reason}"
````

````python
print("Testing the sandbox-->")
fake_good_ai = """ Here is the code!
```python
apples = 5
oranges = 5
result = apples + oranges
```
"""
ans1, err1 = extract_and_run_code(fake_good_ai)
print("TEST1(good code): ")
print(f"Answer: {ans1}")
print(f"Error: {err1}")
fake_bad_ai = """
```python
x = 10
y = 0
result = x / y
```
"""
ans2, err2 = extract_and_run_code(fake_bad_ai)
print("TEST2(bad code): ")
print(f"Answer: {ans2}")
print(f"Error: {err2}")
````

````python
def simulate_agentic_loop(initial_ai_guess, error_fix_guess):
    MAX_RETRIES = 3
    print("\nStarting Agentic Loop Simulation--->")
    print("Attempt 1: AI submits its first code...")
    answer, error = extract_and_run_code(initial_ai_guess)
    if answer is not None:
        print(f"Success on Attempt 1! Answer is {answer}")
        return
    print(f"Attempt 1 Failed! Caught Error: {error}")
    print("Agentic Loop Triggered! Building the 'Retry Prompt'...")
    retry_prompt = f"""
    Your previous code failed with this Sandbox Error:
    {error}
    Rewrite the code to fix the error and find the answer.
    """
    print(f"\n[Sending this back to the AI:]\n{retry_prompt}")
    print("Attempt 2: AI reads the error and submits fixed code...")
    answer2, error2 = extract_and_run_code(error_fix_guess)
    if answer2 is not None:
        print(f"Success on Attempt 2! The Agent fixed its own mistake! Answer is {answer2}")
    else:
        print(f"Attempt 2 Failed Again: {error2}")
first_try_bad_code = """```python
apples = 10
friends = 0
result = apples / friends
```"""
second_try_fixed_code = """```python
apples = 10
friends = 2
result = apples / friends
```"""
````

````python
simulate_agentic_loop(first_try_bad_code, second_try_fixed_code)
````
