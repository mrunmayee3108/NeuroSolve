```python
import pandas as pd
from datasets import load_dataset
```

```python
print("Downloading SVAMP from Hugging Face...")
# We use the 'train' split from HF because SVAMP is a pure challenge dataset,
# but we are tagging it as 'test' in our unified schema!
raw_svamp = load_dataset("ChilleD/SVAMP", split="train")
svamp_df = pd.DataFrame(raw_svamp)
```

```python
print("Transforming SVAMP for the Final Exam...")

clean_svamp = pd.DataFrame()
clean_svamp['id'] = "svp_" + svamp_df['ID'].astype(str)
clean_svamp['dataset_source'] = "SVAMP"
```

```python
# THIS IS CRITICAL: Tagging as 'test' prevents data leakage
clean_svamp['split'] = "test"
```

```python
# SVAMP splits the problem into two columns, so we combine them
clean_svamp['question'] = svamp_df['Body'].str.strip() + " " + svamp_df['Question'].str.strip()
```

```python
# We leave the code EMPTY because the AI has to write it during the test!
clean_svamp['code'] = ""
```

```python
# We keep the ground-truth Answer so we can grade the AI later
clean_svamp['answer'] = svamp_df['Answer']
```

```python
# Ensure exact column order
expected_columns = ['id', 'dataset_source', 'split', 'question', 'code', 'answer']
clean_svamp = clean_svamp[expected_columns]
```

```python
# Save to CSV
output_filename = "unified_svamp_test.csv"
clean_svamp.to_csv(output_filename, index=False)

print(f"\n✅ Success! Saved {len(clean_svamp)} evaluation rows to {output_filename}")
display(clean_svamp.head())
```
