```python
import pandas as pd
from datasets import load_dataset
```

```python
train_df = pd.read_json("train.jsonl", lines=True)
train_df.head()
```

```python
train_df.shape
```

```python
clean_df = pd.DataFrame()
```

```python
clean_df['id'] = "cmq_" + train_df['idx'].astype(str).str.zfill(3)
```

```python
clean_df['dataset_source'] = "codex-math-qa"
clean_df['split'] = "train"
```

```python
clean_df['question'] = train_df['question']
clean_df['code'] = train_df['reply']
clean_df['answer'] = ""
```

```python
clean_df['question'] = clean_df['question'].str.strip()
clean_df['code'] = clean_df['code'].str.strip()
```

```python
clean_df = clean_df[(clean_df['question'] != '') & (clean_df['code'] != '')]
```

```python
initial_rows = len(clean_df)
clean_df = clean_df.drop_duplicates(subset=['question'], keep='first')
dropped_dupes = initial_rows - len(clean_df)
print(f"Removed {dropped_dupes} duplicate questions!")
```

```python
clean_df = clean_df.dropna(subset=['question', 'code'])
```

```python
clean_df = clean_df.head(8000)
```

```python
expected_columns = ['id', 'dataset_source', 'split', 'question', 'code', 'answer']
```

```python
clean_df = clean_df[expected_columns]
```

```python
output_filename = "codex_clean.csv"
clean_df.to_csv(output_filename, index=False)
display(clean_df.head())
```
